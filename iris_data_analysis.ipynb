```python
# Adding Missing Values to data

np.random.seed(42)
df.loc[np.random.choice(df.index,5), "sepal length (cm)"] = np.nan
df.loc[np.random.choice(df.index,3), "target"] = np.nan

df.sample(5)
```

```python
df.columns
```

```python
df["target"] = df["target"].astype("object")
```

```python
df.info()
```

```python
# Creating numeric and categorical coloumns

num_cols = df.select_dtypes(include=["float64"]).columns
num_cols
```

```python
num_cols = df.select_dtypes(include=["float64"]).columns
cat_cols = df.select_dtypes(include=["object"]).columns
```

```python
cat_cols
```

```python
# Repalce the missing value with the median values

for column in num_cols:
  df[column].fillna(df[column].median(),inplace=True)
```

```python
# Converting the categorical data into numerical data

for column in cat_cols:
  le = LabelEncoder()
  df[column] = le.fit_transform(df[column].astype(str))

```

```python
# Normalize data

scaler = MinMaxScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
df.head(3)
```

```python
# Retain only Features with variance > 0.1

selector = VarianceThreshold(threshold=0.05)
num_data = df[num_cols]
select_features = selector.fit_transform(num_data)
```
